```python
import pandas as pd

training = pd.read_csv("/content/breast-w_train.tsv", sep='\t')
display(training.head(n = 10))
```

```python
#will be using support vector machines (SVM) for this classification task
from sklearn import svm
```

A classifier learns the decision boundary used to categorize samples within data.

Different algorithms include linear classifiers, support vector machines or ensembles.

The 2 steps of the classification process are:
1. Building the classifier model. The model learns the data and the manner in which to split inputs and targets into groups

2. Predicting via application of the traind classifier model

```python
brclassifier = svm.LinearSVC()
print(brclassifier)
```

```python
X_train = training.drop('target', axis=1)
y_train = training['target']
brclassifier.fit(X_train, y_train)
```

```python
testing = pd.read_csv("/content/breast-w_test.tsv", sep='\t')
display(testing.head(n = 10))

#X_test = testing.drop('target', axis=1) # Assuming testing also has a 'target' column, if not, remove this line
brclassifier.predict(testing)
```
